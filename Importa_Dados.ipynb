````python
dados.rename(columns={'droga':'composto'}, inplace=True) #muda o nome da coluna, inplace faz com que essa mudança seja feita na tabela original
dados.head()
````

````python
cod_compostos = dados['composto'].value_counts().index[0:5] #aqui estamos vendo quantos tipos de compostos a gente tem e criando um array com os top 5 dos mesmos
cod_compostos
````

````python
dadosCod_compostos = dados.query('composto in @cod_compostos') #filtra a coluna composto, mostrando apenas os top 5 deles
````

````python
import seaborn as sns
sns.set() #utilizar os estilos padroes oferecidos pela seaborn
dadosCod_compostos = dados.query('composto in @cod_compostos')
plt.figure(figsize=(10,8))
ax =sns.countplot(x = 'composto', data = dadosCod_compostos, order = dadosCod_compostos['composto'].value_counts().index) #plota um gráfico, recebe um x,y, também se quiser limitar os dados da coluna como feito acima, em que ele pega apenas os top 5
ax.set_title('Top 5 Compostos', fontsize=20)
plt.show()
````
